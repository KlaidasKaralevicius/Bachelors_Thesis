# Evaluate all models with pycocotools

## Convert YOLO annotations to COCO annotations

```python
import os
import json
from PIL import Image

# convert yolo box annotations to coco annotations
def yolo_to_coco(xc, yc, w, h, img_w, img_h):
    x = (xc - w / 2) * img_w
    y = (yc - h / 2) * img_h
    w = w * img_w
    h = h * img_h
    return [x, y, w, h]

def load_img(img_dir):
    img = []
    img_id_map = {}
    img_id = 1
    
    # skip weird images
    for file in sorted(os.listdir(img_dir)):
        if not file.lower().endswith(('.jpg', '.jpeg')):
            continue
        
        path = os.path.join(img_dir, file)
        
        # get image resolution to get coords for coco
        with Image.open(path) as image:
            w, h = image.size
            
        img.append({
            'id': img_id,
            'file_path': file,
            'width': w,
            'height': h
        })
        
        # connect image to id
        img_id_map[file.replace('.jpg', '').replace('.jpeg', '')] = img_id
        img_id += 1
    return img, img_id_map

# convert ground truth annoations from YOLO to COCO
def convert_labels(label_dir, img, img_id_map):
    annotations = []
    annotation_id = 1
    
    for file in os.listdir(label_dir):
        if not file.endswith('.txt'):
            continue
        
        name = os.path.splitext(file)[0]
        if name not in img_id_map:
            continue
        
        img_id = img_id_map[name]
        # get width/height
        img_info = next(i for i in img if i['id'] == img_id)
        
        with open(os.path.join(label_dir, file)) as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue
                
                parts = line.split()
                if len(parts) != 5:
                    continue
                
                cls, xc, yc, w, h = map(float, parts)
                bbox = yolo_to_coco(xc, yc, w, h, img_info['width'], img_info['height'])

                annotations.append({
                    'id': annotation_id,
                    'image_id': img_id,
                    'category_id': int(cls),
                    'bbox': bbox,
                    'area': bbox[2] * bbox[3],
                    'iscrowd': 0
                })
                annotation_id += 1
    return annotations

# convert predictions from yolo to coco format
def convert_pred(pred_dir, img, img_id_map, class_map=None):
    detections = []
    
    for file in os.listdir(pred_dir):
        if not file.endswith('.txt'):
            continue
        
        name = os.path.splitext(file)[0]
        if name not in img_id_map:
            continue
            
        img_id = img_id_map[name]
        img_info = next(i for i in img if i['id'] == img_id)
        
        with open(os.path.join(pred_dir, file)) as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue
                
                parts = line.split()
                if len(parts) != 6:
                    continue
                
                cls, xc, yc, w, h, conf = map(float, parts)                
                
                bbox = yolo_to_coco(xc, yc, w, h, img_info['width'], img_info['height'])
                
                detections.append({
                    'image_id': img_id,
                    'category_id': int(cls),
                    'bbox': bbox,
                    'score': conf
                })
    return detections
    

def build_categories():
    return [
        {"id": 0, "name": "fork"},
        {"id": 1, "name": "spoon"},
        {"id": 2, "name": "knife"},
        {"id": 3, "name": "ladle"},
        {"id": 4, "name": "spatula"},
        {"id": 5, "name": "bowl"},
        {"id": 6, "name": "plate"},
        {"id": 7, "name": "cup"},
        {"id": 8, "name": "frying pan"},
        {"id": 9, "name": "sauce pan"},
        {"id": 10, "name": "colander"},
        {"id": 11, "name": "pan lid"},
        {"id": 12, "name": "grater"},
        {"id": 13, "name": "cutting board"}
    ]
```

### Convert ground truths for both datasets

```python
# original dataset test subset
img_dir = '../split_dataset/images/test/'
label_dir = '../split_dataset/labels/test/'
# new dataset
img_dir_new = '../new_data_test/images/val/'
label_dir_new = '../new_data_test/labels/val/'

def convert_gt(img_dir, label_dir, output_dir):
    img, img_id_map = load_img(img_dir)
    gt_annotations = convert_labels(label_dir, img, img_id_map)
    categories = build_categories()
    
    gt_json = {
        'images': img, 
        'annotations': gt_annotations,
        'categories': categories
    }
    
    with open(output_dir, 'w') as f:
        json.dump(gt_json, f, indent=2)
        
# original dataset test subset
convert_gt(img_dir, label_dir, '../split_dataset/test_gt.json')
# new test dataset
convert_gt(img_dir_new, label_dir_new, '../new_data_test/gt.json')
```

### Convert model predictions

```python
def convert_model_pred(img_dir, run_dir):
    img, img_id_map = load_img(img_dir)
    
    for model_name in os.listdir(run_dir):
        model_path = os.path.join(run_dir, model_name)
        label_path = os.path.join(model_path, 'labels')
        if not os.path.isdir(label_path):
            continue
        
        preds = convert_pred(label_path, img, img_id_map)
        
        out_file = os.path.join(model_path, 'pred.json')
        
        with open(out_file, 'w') as f:
            json.dump(preds, f, indent=2)

# split_dataset
convert_model_pred(
    img_dir = img_dir,
    run_dir = '../runs/detect/test/'
)
# new_dataset
convert_model_pred(
    img_dir = img_dir_new,
    run_dir = '../runs/detect/new_test/'
)
```

## Evaluate predictions

```python
import os
import json
import contextlib
import numpy as np
from collections import defaultdict

from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval

class_map = {
    42: 7,  # cup
    43: 0,  # fork
    44: 2,  # knife
    45: 1,  # spoon
    46: 5   # bowl
}

datasets = {
    'test': {
        'gt': '../split_dataset/test_gt.json',
        'pred_dir': '../runs/detect/test'
    },
    'new_test': {
        'gt': '../new_data_test/gt.json',
        'pred_dir': '../runs/detect/new_test'
    }
}

model_names = [
    "nano_base", "nano_trained", "nano_trained_openvino",
    "small_base", "small_trained", "small_trained_openvino",
    "medium_base", "medium_trained", "medium_trained_openvino",
    "large_base", "large_trained", "large_trained_openvino",
    "xlarge_base", "xlarge_trained", "xlarge_trained_openvino"
]
```

```python
# silence cocoeval intermediate output
@contextlib.contextmanager
def suppress_stdout():
    with open(os.devnull, "w") as devnull:
        with contextlib.redirect_stdout(devnull):
            yield

# get yolo size and type - base/trained/exported
def parse_model_name(name):
    parts = name.split("_")
    size = parts[0]
    model_type = "_".join(parts[1:])
    return size, model_type

# keep only valid prediction idx (and remap base model predictions)
def remap_predictions(preds, is_base, is_full=True):
    remapped = []
    
    # collect predictions
    for p in preds:
        # if base model, extract only matching pred id's and remap
        if is_base:
            if p['category_id'] not in class_map:
                continue
            
            new_id = class_map[p['category_id']]
        # otherwise just colletc all
        else:
            new_id = p['category_id']
        
        # keep only overlaping idx between all models, instead of full custom dataset idx
        if not is_full:
            overlap_class_ids = set(class_map.values())
            if new_id not in overlap_class_ids:
                continue
        
        # shift by 1 for cocoeval tool
        new_id += 1
        
        new_p = p.copy()
        new_p['category_id'] = new_id
        remapped.append(new_p)
    return remapped

# shift ground truth idx by 1 for cocoeval tool
def shift_gt(coco_gt, is_full=True):
    # just push all idx by 1 for full class evaluation
    if is_full:
        for ann in coco_gt.dataset['annotations']:
            ann['category_id'] += 1
        
        for cat in coco_gt.dataset['categories']:
            cat['id'] += 1
    # push idx for only overlaping class idx
    else:
        # find overlapping idx
        overlap_class_ids = set(class_map.values())
        overlap_class_ids_shifted = {c + 1 for c in overlap_class_ids}
        new_anns = []
        
        # shift by 1 only overlaping
        for ann in coco_gt.dataset['annotations']:
            new_id = ann['category_id'] + 1
            if new_id in overlap_class_ids_shifted:
                ann['category_id'] = new_id
                new_anns.append(ann)
        
        coco_gt.dataset['annotations'] = new_anns
        
        # shift by 1 only overlaping
        new_cats = []
        for cat in coco_gt.dataset['categories']:
            new_id = cat['id'] + 1
            if new_id in overlap_class_ids_shifted:
                cat['id'] = new_id
                new_cats.append(cat)
        
        coco_gt.dataset['categories'] = new_cats
        
    coco_gt.createIndex()

# get model and per-class metrics
def eval_metrics(gt_path, pred_path, is_base, is_full=True):
    # load and shift ground truth
    with suppress_stdout():
        coco_gt = COCO(gt_path)
        shift_gt(coco_gt, is_full)
    
    # get overlaping idx if not evaluation on 14 classes
    if not is_full:
        overlap_class_ids = set(class_map.values())
        overlap_class_ids_shifted = {c + 1 for c in overlap_class_ids}
    
    # read and remap prediction json
    with open(pred_path, 'r') as f:
        preds = json.load(f)
    preds = remap_predictions(preds, is_base, is_full)
    
    if len(preds) == 0:
        return None
    
    # choose to evaluate on 14 vs 5 classes
    if is_full:
        valid_cat_ids = [cat['id'] for cat in coco_gt.dataset['categories']]
    else:
        valid_cat_ids = list(overlap_class_ids_shifted)
        
    # run cocoeval tools for metrics
    with suppress_stdout():
        coco_dt = coco_gt.loadRes(preds)
        
        coco_eval = COCOeval(coco_gt, coco_dt, 'bbox')
        
        coco_eval.params.catIds = valid_cat_ids
        coco_eval.params.imgIds = list(set(p['image_id'] for p in preds))
        
        coco_eval.evaluate()
        coco_eval.accumulate()
        coco_eval.summarize()
        
    stats = coco_eval.stats
    
    metrics = {
        'AP': stats[0],
        'AP50': stats[1],
        'AP75': stats[2],
        'AR': stats[8]
    }
    
    # per class ap (ap@0.5-0.95)
    precisions = coco_eval.eval['precision']
    cat_ids = coco_eval.params.catIds
    per_class_ap = {}
    
    # get names instead of idx
    cat_id_to_name = {c['id']: c['name'] for c in coco_gt.dataset['categories']}
    
    # extract metrics
    for idx, cat_id in enumerate(cat_ids):
        # remove invalid entries
        precision = precisions[:, :, idx, 0, -1]
        precision = precision[precision > -1]
        
        # compute ap
        if len(precision) == 0:
            ap = float('nan')
        else:
            ap = np.mean(precision)
        
        # map id to class name
        name = cat_id_to_name.get(cat_id, str(cat_id))
        per_class_ap[name] = ap
        
    return metrics, per_class_ap
```

```python
def print_table(dataset_name, data):
    print(f"\n=== DATASET: {dataset_name} ===")

    for size in ["nano", "small", "medium", "large", "xlarge"]:
        if size not in data:
            continue

        print(f"\n--- {size.upper()} ---")

        row = data[size]

        for metric in ["AP", "AP50", "AP75", "AR"]:
            base = row.get("base", {}).get(metric, 0)
            trained = row.get("trained", {}).get(metric, 0)
            ov = row.get("trained_openvino", {}).get(metric, 0)

            print(
                f"{metric:5} | "
                f"base: {base:.5f} | "
                f"trained: {trained:.5f} | "
                f"openvino: {ov:.5f}"
            )


def print_per_class(dataset_name, data, size="nano"):
    print(f"\n=== PER-CLASS ({dataset_name}, {size}) ===")

    row = data.get(size, {})

    base = row.get("base", {})
    trained = row.get("trained", {})
    ov = row.get("trained_openvino", {})

    classes = sorted(set(base) | set(trained) | set(ov))

    for cls in classes:
        print(
            f"{cls:15} | "
            f"{base.get(cls, 0):.4f} | "
            f"{trained.get(cls, 0):.4f} | "
            f"{ov.get(cls, 0):.4f}"
        )
```

```python
def evaluate(is_full=True):
    results = defaultdict(lambda: defaultdict(dict))
    per_class_results = defaultdict(lambda: defaultdict(dict))
    
    for dataset_name, dataset_info in datasets.items():
        gt_path = dataset_info['gt']
        pred_dir = dataset_info['pred_dir']
        
        for model_name in model_names:
            size, model_type = parse_model_name(model_name)
            
            pred_path = os.path.join(pred_dir, f'{model_name}/pred.json')
            
            if not os.path.exists(pred_path):
                print(f'Missing: {pred_path}')
                continue
            
            is_base = 'base' in model_type
            
            out = eval_metrics(gt_path, pred_path, is_base, is_full)
            
            if out is None:
                continue
            
            metrics, per_class_ap = out
            
            results[dataset_name][size][model_type] = metrics
            per_class_results[dataset_name][size][model_type] = per_class_ap  
    
    for dataset_name, data in results.items():
        print_table(dataset_name, data)
    
    for dataset_name, data in per_class_results.items():
        for size in ['nano', 'small', 'medium', 'large', 'xlarge']:
            if size in data:
                print_per_class(dataset_name, data, size=size)
    
    return results, per_class_results
```

## Raspberry Pi resource evaluation

```python
import csv
import numpy as np

def fmt(avg, p95):
    return f"{avg:.2f} / {p95:.2f}"

def analyze_file(path, power_avg=None, power_max=None):
    latency = []
    cpu_avg = []
    cpu_max = []
    freq_avg = []
    
    ram_used = []
    temp_max = []
    
    throttle_count = 0

    with open(path, 'r') as f:
        reader = csv.DictReader(f)
        
        for row in reader:
            latency.append(float(row['latency_ms']))
            cpu_avg.append(float(row['cpu_avg']))
            cpu_max.append(float(row['cpu_max']))
            freq_avg.append(float(row['freq_avg']))
            
            ram_used.append(float(row['ram_used']))
            temp_max.append(float(row['temp_max']))
            
            throttle_count += int(row['throttle_count'])

    # averages
    latency_avg = np.mean(latency)
    cpu_avg_avg = np.mean(cpu_avg)
    cpu_max_avg = np.mean(cpu_max)
    freq_avg_avg = np.mean(freq_avg)

    temp_avg = np.mean(temp_max)

    results = {
        'latency_ms': fmt(latency_avg, np.percentile(latency, 95)),
        'cpu_avg': f'{fmt(cpu_avg_avg, np.percentile(cpu_avg, 95))}%',
        'cpu_max': f'{fmt(cpu_max_avg, np.percentile(cpu_max, 95))}%',
        'freq_avg': fmt(freq_avg_avg, np.percentile(freq_avg, 95)),

        'ram_used_max': f'{max(ram_used)}GB',

        # changed: temp → power format
        'temp': f"{temp_avg:.2f}C / {max(temp_max):.2f}C",

        'throttle_count': throttle_count
    }

    # manual power input
    if power_avg is not None and power_max is not None:
        results['power'] = f"{power_avg:.2f}W / {power_max:.2f}W"

    return results
```

```python
def show_resources(path, p_avg, p_max):
    stats = analyze_file(
        path = path,
        power_avg = p_avg,
        power_max = p_max
        )

    print(f'==={os.path.basename(path)}===')
    print('--Average / p95--')
    for k, v in stats.items():
        print(f'{k}: {v}')
    print('\n')
    return stats
```
