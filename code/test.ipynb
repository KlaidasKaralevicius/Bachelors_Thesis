```python
from ultralytics import YOLO
model = YOLO("weights/yolo26m-trained.pt")  # yolo26[n-s-m-l-x].pt yolo26[n-s-m-l-x]-trained.pt
```

```python
results = model.val(
    data='dataset.yaml',
    split='test',
    plots=True,
    project='test',
    name='medium-trained', # nano, small, medium, large, xlarge | nano-trained small-trained medium-trained large-trained xlarge-trained
    seed=226,
    exist_ok=True,
    device=0
)
```
