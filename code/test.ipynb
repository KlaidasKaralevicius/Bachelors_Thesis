### Extract only predictions of known classes from base models

```python
import json
# custom classes appearing in COCO
coco_to_custom = {
    7: 42, # cup
    0: 43, # fork
    2: 44, # knife
    1: 45, # spoon
    5: 46  # bowl
}

def filter_base_val(input_path, output_path):
    input_path = input_path
    output_path = output_path
    
    with open(input_path, 'r') as f:
        preds = json.load(f)
        
    remapped = []
    for p in preds:
        coco_cls = p['category_id']
        if coco_cls in coco_to_custom:
            p['category_id'] = coco_to_custom[coco_cls]
            remapped.append(p)
    
    with open(output_path, 'w') as f:
        json.dump(remapped, f)
```
